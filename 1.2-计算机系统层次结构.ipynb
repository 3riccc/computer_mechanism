# 计算机系统层次结构

<hr/>
## 组成
- 硬件系统
- 软件系统


- 软硬件在逻辑等效
    - 意思是一个功能可以用软件实现也可以用硬件实现
    - 硬件效率高
    - 软件灵活性高

<hr/>
## 硬件基本组成

### 早期冯诺伊机
- “存储程序”的思想奠定了现代计算机基本结构
- 按照这个思路做出来的机器都叫冯诺伊曼机




1. 硬件由运算器、存储器、控制器、输入设备、输入设备5大部分组成

1. 指令和数据以同等地位存储在存储器内，可以按地址访问

1. 指令和数据均用二进制代码表示

1. 指令包括
    1. 操作码：表示操作是干啥
    1. 地址码：表示操作数在哪

1. 指令在存储器内顺序存放。通常也顺序执行，在特定条件下，执行顺序可以改变

1. 运算器是中心




### 现代计算机组织结构

**不以处理器为中心，而以存储器为中心**

- I/O操作尽可能绕过CPU，直接在I/O设备和存储器之间完成
- 可以提高效率

## 功能部件——存储器

- 存放程序和数据

- 分成主存（也称内存）和辅助存储器（也称外存）

- CPU能直接访问的是主存，外存的信息必须先调入主存，才能被CPU访问

- 主存
    - 分成若干个存储单元
        1. 分成若干个存储元件
            - 每个存储元件存储一位1或0
        1. 所以每个存储单元存储一串二进制代码：这串代码称为**存储字**
        1. 这串代码的长度：**存储字长**
        1. 存储字长可以是8bit的偶数倍

- 存储方式：按存储单元地址进行存取

- 组成如图：
    


- 上图解读
    - 存储体存放正式信息
    - 地址寄存器（**MAR**）存放访存地址，经过地址译码之后，找到所选的存储单元
    - 数据寄存器（**MDR**）主存和其他部件中介机构，暂存要从存储器中读或者写的信息
    
- MAR（地址寄存器）：
    - 用于在主存中寻址
    - 位数对应存储单元个数
        - 10位，说明主存中有1024个存储单元
- MDR（数据寄存器）：
    - 在现代计算机中，MDR存在于CPU中，另外，高速缓存Cache也存在于CPU中

<hr/>
## 功能部件——运算器

完成算术运算（加减乘除）和逻辑运算（与或非）

- 核心是算术逻辑单元ALU（Arithmetic Logical Unit）

<hr/>
## 功能部件——控制器

指挥各部门协调合作

包括
- 程序计数器（PC）：存放当前指令地址，可以自动+1，与主存的MAR之间有一个直接通路
- 指令寄存器（IR）：
    - 存放当前指令，内容来自主存的MDR
    - 指令中的操作码送给CU，CU会分析指令和发出微操作命令
    - 指令中的地址码送给MAR，读取数据
- 控制单元（CU）


**现代计算机一般把运算器和控制器写到一个芯片里，合称CPU**

## 计算机软件分类

- 系统软件（记这个感觉就得了）：
    - 操作系统
    - 数据库管理系统（这个和我们写代码用的数据库不一样，这个是用户和操作系统之间的管理软件，可能是用来管理系统配置数据或者什么的吧，是系统软件）
    - 语言处理程序
    - 分布式软件系统
    - 网络软件系统
    - 标准程序库
    - 服务性程序
   

- 应用软件
    - 略
   

- 机器语言
    - 二进制代码
    - 是计算机唯一可以直接执行的语言

- 汇编语言
    - 英文单词或者缩写代替二进制指令代码
    - 必须经过汇编程序（一个系统软件）的翻译转化为机器语言才能执行
    
- 高级语言
    - C、java等
    - 经过编译程序编译成汇编语言
    - 再经过汇编程序编译成机器语言程序，才能执行
